```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import csv
import numpy as np
```

```python
def running_mean(x, N=36):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
mpl.style.use('default')
cmap = plt.get_cmap("tab10")
```

```python
nn_v0_batt = np.load("./all_batt_2017_NN-v0.npy")
drl_v0_batt = np.load("./DRL-v0_node_batt_rec.npy")
drl_v0_explore_batt =np.load("./DRL-v0_explore_node_batt_rec.npy")
drl_v0_cp_batt =np.load("./DRL_CP-v0_node_batt_rec.npy")
drl_v0_explore_cp_batt =np.load("./DRL_CP_explore-v0_node_batt_rec.npy")

fig = plt.figure(figsize=(10,20))
# fig.subplots_adjust(hspace=2)
fig.suptitle("BATTERY PROFILES for ENO-V0", size = 25)
# ax0 = fig.add_subplot(5,1,1)
# ax0.set_title("NN", size = 20)
# ax0.plot(running_mean(nn_v0_batt),alpha = 1.0,linewidth=0.7,label="NN")
# ax0.set_ylim([0,1])
# ax0.set_ylabel("Normalized Battery", size = 16)
# ax0.set_xlabel("Hours", size = 16)
# ax0.legend(loc = "upper right",ncol=5)



ax1 = fig.add_subplot(5,1,1)
ax1.set_title("\n\nDRL", size = 20)
for i in range(drl_v0_batt.shape[0]):
    ax1.plot(running_mean(drl_v0_batt[i][:150*24],100),alpha = 1.0,linewidth=0.7,label="Node"+str(i))
ax1.plot(running_mean(nn_v0_batt[:150*24],100),alpha = 0.6,linewidth=1.5,label="NN",color='k',linestyle=":")
ax1.set_ylim([0,1])
ax1.set_ylabel("Normalized Battery", size = 16)
ax1.set_xlabel("Hours", size = 16)
ax1.legend(ncol=6,loc="best")

ax2 = fig.add_subplot(5,1,2)
ax2.set_title("DRL-XPLR", size = 20)
for i in range(drl_v0_batt.shape[0]):
    ax2.plot(running_mean(drl_v0_explore_batt[i][:150*24],100),alpha = 1.0,linewidth=0.7,label="Node"+str(i))
ax2.plot(running_mean(nn_v0_batt[:150*24],100),alpha = 0.6,linewidth=1.5,label="NN",color='k',linestyle=":")
ax2.set_ylim([0,1])
ax2.set_ylabel("Normalized Battery", size = 16)
ax2.set_xlabel("Hours", size = 16)
# ax2.legend(loc = "upper right",ncol=6)


ax3 = fig.add_subplot(5,1,3)
ax3.set_title("DRL-CP", size = 20)
for i in range(drl_v0_batt.shape[0]):
    ax3.plot(running_mean(drl_v0_cp_batt[i][:150*24],100),alpha = 1.0,linewidth=0.7,label="Node"+str(i))
ax3.plot(running_mean(nn_v0_batt[:150*24],100),alpha = 0.6,linewidth=1.5,label="NN",color='k',linestyle=":")
ax3.set_ylim([0,1])
ax3.set_ylabel("Normalized Battery", size = 16)
ax3.set_xlabel("Hours", size = 16)
# ax3.legend(loc = "upper right",ncol=6)


ax4 = fig.add_subplot(5,1,4)
ax4.set_title("DRL-XCP", size = 20)
for i in range(drl_v0_batt.shape[0]):
    ax4.plot(running_mean(drl_v0_explore_cp_batt[i][:150*24],100),alpha = 1.0,linewidth=0.7,label="Node"+str(i))
ax4.plot(running_mean(nn_v0_batt[:150*24],100),alpha = 0.6,linewidth=1.5,label="NN",color='k',linestyle=":")
ax4.set_ylim([0,1])
ax4.set_ylabel("Normalized Battery", size = 16)
ax4.set_xlabel("Hours", size = 16)
# ax4.legend(loc = "upper right",ncol=6)


fig.tight_layout()
plt.show()
```
